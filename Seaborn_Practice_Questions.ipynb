## Question 1: Comparing Two Lists Using a Line Plot

Problem Statement:
You are given two lists representing monthly sales of two products over 6 months.
You need to visualize how their sales trends compare over time using a line plot.

Approach:
1. Create two lists: months and sales.
2. Use `sns.lineplot()` to plot both lines on the same graph.
3. Add appropriate labels and a title to make the visualization clear.

```python
# Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Example lists
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_A = [45, 55, 60, 70, 65, 75]
sales_B = [40, 50, 58, 68, 70, 80]

# Your line plot code here

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Product_A': [150, 200, 250, 300, 280, 350],
    'Product_B': [180, 220, 210, 260, 300, 330]
}

df = pd.DataFrame(data)
df_melted = pd.melt(df, id_vars=['Month'], value_vars=['Product_A', 'Product_B'],
                    var_name='Product', value_name='Sales')

sns.lineplot(data=df_melted, x='Month', y='Sales', hue='Product', marker='o')
plt.title('Monthly Sales Comparison (Product A vs Product B)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

## Question 3: Distribution of Total Fare Amounts

Problem Statement:
We want to understand how total fare amounts are distributed among taxi trips.

Approach:
1. Use the 'fare' column from the 'taxis' dataset.

## Question 8: Trip Duration vs. Fare Amount

Problem Statement:
Is there any relationship between how long a trip takes and how much it costs?

Approach:
1. Use columns like 'pickup' and 'dropoff' to calculate trip duration (if timestamps are available).
2. Plot a scatter plot between duration and fare using `sns.scatterplot()`.

```python
# Write your code here
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

taxis = sns.load_dataset('taxis')

taxis['pickup'] = pd.to_datetime(taxis['pickup'])
taxis['dropoff'] = pd.to_datetime(taxis['dropoff'])
taxis['duration'] = (taxis['dropoff'] - taxis['pickup']).dt.total_seconds() / 60
sns.scatterplot(data=taxis, x='duration', y='fare', color='purple', alpha=0.6)

plt.title("Trip Duration vs. Fare Amount", fontsize=14)
plt.xlabel("Trip Duration (minutes)", fontsize=12)
plt.ylabel("Fare Amount ($)", fontsize=12)

plt.show()
```

## Question 10: Average Tip by Distance Category

Problem Statement:
We want to know if longer trips generally result in higher tips.

Approach:
1. Create distance categories (e.g., short, medium, long) using the 'distance' column.
2. Use `sns.barplot()` to plot average tips for each distance category.

```python
# Write your code here
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

taxis = sns.load_dataset('taxis')

bins = [0, 2, 5, taxis['distance'].max()]
labels = ['Short', 'Medium', 'Long']
taxis['distance_category'] = pd.cut(taxis['distance'], bins=bins, labels=labels)
sns.barplot(data=taxis, x='distance_category', y='tip', palette='viridis', estimator='mean')

plt.title("Average Tip by Distance Category", fontsize=14)
plt.xlabel("Distance Category", fontsize=12)
plt.ylabel("Average Tip Amount ($)", fontsize=12)

plt.show()
```
