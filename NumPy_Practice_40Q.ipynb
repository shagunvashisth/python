### Q38. Using NumPy, compute each student’s **total score** (math + english + science) and print the `id` of the student with the highest total.

```python
# answer here
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'math': [78, 85, 92, 66, 74],
    'english': [81, 79, 88, 70, 77],
    'science': [75, 90, 85, 68, 80]
})

math_array = df['math'].to_numpy()
english_array = df['english'].to_numpy()
science_array = df['science'].to_numpy()

total_scores = math_array + english_array + science_array

top_index = np.argmax(total_scores)
top_student_id = df['student_id'].iloc[top_index]

print("Total Scores:", total_scores)
print("Student with highest total score is ID:", top_student_id)
print("Highest Total Score:", total_scores[top_index])
```

### Q39. Using NumPy, standardize the `math` scores to **z-scores**: `(x - mean) / std` and print the first 5 standardized values.

```python
# answer here
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'math': [78, 85, 92, 66, 74],
    'english': [81, 79, 88, 70, 77],
    'science': [75, 90, 85, 68, 80]
})

math_array = df['math'].to_numpy()

mean_math = np.mean(math_array)
std_math = np.std(math_array)
z_scores = (math_array - mean_math) / std_math

print("Mean of math scores:", mean_math)
print("Standard deviation:", std_math)
print("First 5 standardized (z-score) values:")
print(z_scores[:5])
```

### Q40. Using NumPy logical operations, count how many students scored **≥ 70 in all three subjects** and print that count.

```python
# answer here
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'math': [78, 85, 92, 66, 74],
    'english': [81, 79, 88, 70, 77],
    'science': [75, 90, 85, 68, 80]
})

math_array = df['math'].to_numpy()
english_array = df['english'].to_numpy()
science_array = df['science'].to_numpy()

condition = (math_array >= 70) & (english_array >= 70) & (science_array >= 70)
count = np.sum(condition)

print("Number of students scoring ≥ 70 in all subjects:", count)
```
